```python
import json
import pandas as pd

# List your file paths
file_paths = [
    '/content/compoundV2_transactions_ethereum_chunk_13 (1).json',
    '/content/compoundV2_transactions_ethereum_chunk_18.json',
    '/content/compoundV2_transactions_ethereum_chunk_2.json'
]

# Action types you want to extract
action_types = ['deposits', 'withdraws', 'borrows', 'repays', 'liquidates']

# Initialize empty lists to hold DataFrames per action type
global_dfs = {action: [] for action in action_types}

# Loop through each file and extract action data
for file_path in file_paths:
    with open(file_path, 'r') as f:
        data = json.load(f)

    for action in action_types:
        if action in data:
            if data[action]:  # Only append if action has non-empty data
                df = pd.json_normalize(data[action])
                global_dfs[action].append(df)
            else:
                print(f"'{action}' is present in {file_path} but contains no data.")
        else:
            print(f"'{action}' not found in {file_path}.")

# Safely concatenate non-empty action DataFrames
final_dfs = {}
for action in action_types:
    if global_dfs[action]:  # Only concatenate if there's at least one DataFrame
        final_dfs[action] = pd.concat(global_dfs[action], ignore_index=True)
        print(f"{action}: Loaded {len(final_dfs[action])} rows.")
    else:
        print(f"No data found for action: {action}")

```

```python
# Example usage:
df_deposits = final_dfs['deposits']
df_withdraws = final_dfs['withdraws']
df_borrows = final_dfs['borrows']
df_repays = final_dfs['repays']
df_liquidates = final_dfs['liquidates']


```

aggregating per wallet (i.e., account.id) across all 4 datasets:

Total USD amount deposited

Total USD amount borrowed

Total USD amount repaid

Total USD amount withdrawn

Number of transactions of each type

Active time period (duration between first and last transaction)

```python
deposits_agg = df_deposits.groupby('account.id').agg(total_deposits_usd =('amountUSD', 'sum'),
                                                     deposit_tx_count=('amountUSD'	,'count'),
                                                     first_deposit=('timestamp','min'),
                                                     last_deposit=('timestamp','max')).reset_index()
```

```python
deposits_agg.head()
```

```python
withdraws_agg = df_withdraws.groupby('account.id').agg(total_withdraws_usd =('amountUSD', 'sum'),
                                                     withdraws_tx_count=('amountUSD'	,'count'),
                                                     first_withdraws=('timestamp','min'),
                                                     last_withdraws=('timestamp','max')).reset_index()
```

```python
withdraws_agg.head()
```

```python
borrows_agg = df_borrows.groupby('account.id').agg(total_borrows_usd =('amountUSD', 'sum'),
                                                     borrows_tx_count=('amountUSD'	,'count'),
                                                     first_borrows=('timestamp','min'),
                                                     last_borrows=('timestamp','max')).reset_index()
```

```python
borrows_agg.head()
```

```python
repays_agg = df_repays.groupby('account.id').agg(total_repays_usd =('amountUSD', 'sum'),
                                                     repays_tx_count=('amountUSD'	,'count'),
                                                     first_repays=('timestamp','min'),
                                                     last_repays=('timestamp','max')).reset_index()
```

```python
repays_agg.head()
```

```python
liquidates_agg = (
    df_liquidates.groupby('account.id')
    .agg(
        total_liquidations_usd=('amountUSD', 'sum'),
        liquidation_tx_count=('amountUSD', 'count'),
        first_liquidation=('timestamp', 'min'),
        last_liquidation=('timestamp', 'max')
    )
    .reset_index()
)
```

```python
liquidates_agg.head()
```

```python
from functools import reduce

dfs = [deposits_agg, withdraws_agg, borrows_agg, repays_agg,liquidates_agg]
df_all = reduce(lambda left, right: pd.merge(left, right, on='account.id', how='outer'), dfs)
```

```python
# Only fill count columns with 0 (makes logical sense)
count_cols = [
    'deposit_tx_count', 'withdraws_tx_count',
    'borrows_tx_count', 'repays_tx_count',
    'liquidation_tx_count'
]
df_all[count_cols] = df_all[count_cols].fillna(0)
```

1. Net Deposit (Deposits  Withdrawals)

```python
df_all['net_deposit_usd'] = (
    df_all['total_deposits_usd'].fillna(0) - df_all['total_withdraws_usd'].fillna(0)
)
```

 Net Borrow (Borrows  Repays)

```python
df_all['net_borrow_usd'] = (
    df_all['total_borrows_usd'].fillna(0) - df_all['total_repays_usd'].fillna(0)
)
```

Active Duration  how long an account was active (min and max across all timestamps):

```python
df_all['first_action'] = df_all[['first_deposit', 'first_withdraws', 'first_borrows', 'first_repays']].min(axis=1)
df_all['last_action'] = df_all[['last_deposit', 'last_withdraws', 'last_borrows', 'last_repays']].max(axis=1)
df_all['active_days'] = (df_all['last_action'] - df_all['first_action']).dt.days
```

Total Transactions Count

```python
df_all['total_tx_count'] = (
    df_all['deposit_tx_count'].fillna(0) +
    df_all['withdraws_tx_count'].fillna(0) +
    df_all['borrows_tx_count'].fillna(0) +
    df_all['repays_tx_count'].fillna(0) +
    df_all['liquidation_tx_count'].fillna(0)
)
```

Total Volume

```python
df_all['total_volume_usd'] = (
    df_all['total_deposits_usd'].fillna(0) +
    df_all['total_withdraws_usd'].fillna(0) +
    df_all['total_borrows_usd'].fillna(0) +
    df_all['total_repays_usd'].fillna(0) +
    df_all['total_liquidations_usd'].fillna(0)
)
```

Average Transaction value

```python
df_all['avg_tx_value_usd'] = df_all['total_volume_usd'] / df_all['total_tx_count'].replace(0, np.nan)

```

deposit to borrow ratio

```python
df_all['deposit_to_borrow_ratio'] = df_all['total_borrows_usd'] / df_all['total_deposits_usd'].replace(0, np.nan)

```

repay_ratio

```python
df_all['repay_ratio'] = df_all['total_repays_usd'] / df_all['total_borrows_usd'].replace(0, np.nan)

```

liquidation_rate

```python
df_all['liquidation_rate'] = df_all['total_liquidations_usd'] / df_all['total_borrows_usd'].replace(0, np.nan)
```

Scoring function

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Define the features and raw weights
features = [
    'net_deposit_usd', 'net_borrow_usd', 'active_days',
    'total_volume_usd', 'deposit_to_borrow_ratio',
    'repay_ratio', 'liquidation_rate'
]

weights = np.array([0.2, 0.1, 0.15, 0.15, 0.15, 0.15, -0.1])

# Copy the DataFrame and fill NaNs
df_score = df_all.copy()
df_score[features] = df_score[features].fillna(0)

# Scale the features (0 to 1)
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(df_score[features])

# Compute raw weighted score
df_score['score_raw'] = (scaled_features * weights).sum(axis=1)

# Rescale raw score to 0100
df_score['score'] = MinMaxScaler().fit_transform(df_score[['score_raw']]) * 100

# Drop intermediate column if needed
df_score.drop(columns=['score_raw'], inplace=True)
```

```python
df_score.head()
```

```python
def assign_label(score):
    if score >= 80:
        return 2  # Highly Trustable
    elif score >= 50:
        return 1  # Medium Trust
    else:
        return 0  # High Risk
df_score['label'] = df_score['score'].apply(assign_label)
```

```python
df_sorted = df_score.sort_values(by='score', ascending=False)
```
